# ALGORITMO: vecinos comunes

```python
def alg_common_neighbors(G, GTest):
    GScoreRight = nx.Graph() #guardar las puntuaciones correcta
    GScoreWrong = nx.Graph() #guardar las puntuaciones incorrectas
    nodeList = list(G.nodes())
    
    startTime = time.time() #BEGIN
    
    for n in nodeList: #para todos los nodos de la red
        for nn in nx.non_neighbors(G,n): #comparamos con aquellos que no tienen conexion
            scoreCN = len(sorted(nx.common_neighbors(G,n,nn)))
            if scoreCN > 0  and not GScoreRight.has_edge(n,nn) and not GScoreWrong.has_edge(n,nn):
                if GTest.has_edge(n,nn):
                    GScoreRight.add_edge(n,nn,score=scoreCN)
                else:
                    GScoreWrong.add_edge(n,nn,score=scoreCN)
     
    finishTime = time.time() - startTime #END
    return GScoreRight, GScoreWrong, finishTime
```

```python
def lp_AUC(GScoreRight, GScoreWrong, n):
    #validamos el valor de N
    if n > GScoreRight.number_of_edges():
        n = int(GScoreRight.number_of_edges() * 0.9)
    
    #seleccionamos aleatoriamente los elementos right y wrong
    lisR = list(GScoreRight.edges())
    lis_r = list(range(len(lisR)))
    shuffle(lis_r)
    lisW = list(GScoreWrong.edges())
    lis_w = list(range(len(lisW)))
    shuffle(lis_w)
    
    ni = 0 #numero de veces que un link correcto tiene mejor score que un link incorrecto
    nii = 0  #numero de veces que un link correcto tiene igual score que un link incorrecto
    
    for i in range(0, n):
        if GScoreRight.get_edge_data(*lisR[lis_r[i]])['score'] > GScoreWrong.get_edge_data(*lisW[lis_w[i]])['score']:
            ni += 1
        else:
            if GScoreRight.get_edge_data(*lisR[lis_r[i]])['score'] == GScoreWrong.get_edge_data(*lisW[lis_w[i]])['score']:
                nii += 1
    
    AUC = (ni + (nii * 0.5)) / n
    return AUC
```

```python
def lp_measures(ALG, GTest, GScoreRight, GScoreWrong, finishTime):
    PRECISION = GScoreRight.number_of_edges()/GTest.number_of_edges()
    TIEMPO = finishTime
    AUC = lp_AUC(GScoreRight, GScoreWrong, 100)
    
    print(ALG+":")
    print("\n")
    print("  PRECISION: "+str(PRECISION)) 
    print("\n")
    print("  TIEMPO: "+str(TIEMPO)) 
    print("\n")
    print("  AUC: "+str(AUC))
    print("\n")
```

```python
def train_test_graph(links, p): #pandas con los links y porcentaje entre 0 y 1 de lo que se desea en el test
    if p > 1 or p < 0:
        p = 0.1
        
    links = shuffle(links) #mezclar aleatoriamente los links
    N = len(links) #numero de links
    N_train = N - N * p
    
    G_train = nx.Graph()
    G_test  = nx.Graph()
    
    
    for i, elrow in links.iterrows():
        if i < N_train:
            G_train.add_edge(elrow[0], elrow[1], attr_dict=elrow[2:].to_dict())
        else:
             G_test.add_edge(elrow[0], elrow[1], attr_dict=elrow[2:].to_dict())

    return G_train, G_test
```

```python
G_train, G_test = train_test_graph(links, 0.1)

GScoreRight, GScoreWrong, finishTime = alg_common_neighbors(G_train, G_test)
lp_measures("Vecinos Comunes", G_test, GScoreRight, GScoreWrong, finishTime)
```
